```python
# 1. ELECCIÓN Y CARGA DE DATOS

# Elegí un dataset de Kaggle sobre trabajos de Supermarket Sales
# Archivo: Supermarket Sales 2.csv
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

uploaded = files.upload()

nombre_archivo = list(uploaded.keys())[0]
print("Archivo subido:", nombre_archivo)

# CARGA DEL CSV

df = pd.read_csv(nombre_archivo)
print("✓ Dataset cargado correctamente\n")
```

```python
# Muestra las primeras 5 filas (head())
print("--- Primeras 5 filas (df.head()) ---")
print(df.head())

# Muestra las últimas 5 filas (tail())
print("\n--- Últimas 5 filas (df.tail()) ---")
print(df.tail())

# Muestra el número de filas y columnas (shape)
print("\n--- Forma del DataFrame (df.shape) ---")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

# Muestra información de los tipos de datos, valores no nulos y uso de memoria (info())
print("\n--- Información del DataFrame (df.info()) ---")
df.info()

# Muestra estadísticas descriptivas para columnas numéricas (describe())
print("\n--- Estadísticas Descriptivas (df.describe()) ---")
print(df.describe())
```

```python
import numpy as np

# Reemplazar 'Unknown' en Episodes por NaN para manejarlo numéricamente
df['Episodes'] = df['Episodes'].replace('Unknown', np.nan)

# Convertir la columna Episodes a tipo entero, ignorando los NaN
df['Episodes'] = pd.to_numeric(df['Episodes'])

# Mostrar cuántos valores nulos hay ahora en Episodes
print(f"Valores nulos en 'Episodes' después de la limpieza: {df['Episodes'].isna().sum()}")
```

```python
# Conteo de frecuencia de los valores en la columna 'Studios'
top_10_estudios = df['Studios'].value_counts().head(10)

print("\n--- Top 10 Estudios con más Animes Producidos ---")
print(top_10_estudios)
```

```python
# Usamos idxmax() en la columna 'Episodes' limpiada para encontrar el índice del valor máximo
indice_max_episodios = df['Episodes'].idxmax()

# Obtenemos el título y la cantidad de episodios de esa fila
anime_mas_largo = df.loc[indice_max_episodios, ['title', 'Episodes']]

print("\n--- Anime con el Mayor Número de Episodios ---")
print(f"Título: {anime_mas_largo['title']}")
print(f"Episodios: {int(anime_mas_largo['Episodes'])}")
```

```python

def obtener_temporada(premiered):
    if pd.isna(premiered) or premiered == 'Unknown':
        return 'N/A'
    # La temporada es la primera palabra
    return premiered.split(' ')[0]

# Aplicar la función para crear la nueva columna 'Season'
df['Season'] = df['Premiered'].apply(obtener_temporada)

print("\n--- Vista previa de la nueva columna 'Season' ---")
print(df[['title', 'Premiered', 'Season']].head())
```
