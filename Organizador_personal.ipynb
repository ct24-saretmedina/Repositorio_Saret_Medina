```python
print("=== ORGANIZADOR DE POKÉMON  ===")

pokemones = []  # Lista principal

while True:
    print("\nMenú de opciones:")
    print("1. Ver pokemones")
    print("2. Agregar un Pokémon")
    print("3. Eliminar un Pokémon")
    print("4. Actualizar información de un Pokémon")
    print("5. Salir")

    opcion = input("Elige una opción: ")

    # 1. Ver lista
    if opcion == "1":
        if pokemones:
            print("\n Tu equipo de Pokémon:")
            for i, p in enumerate(pokemones):
                print(f"{i+1}. {p[0]}  | Tipo: {p[1]} | Nivel: {p[2]}")
        else:
            print("\nNo tienes pokemones registrados.")

    # 2. Agregar Pokémon
    elif opcion == "2":
        nombre = input("Nombre del Pokémon: ")
        tipo = input("Tipo del Pokémon (Agua, Fuego, Planta, etc.): ")
        nivel = int(input("Nivel del Pokémon: "))
        pokemones.append([nombre, tipo, nivel])
        print(" Pokémon agregado a tu equipo.")

    # 3. Eliminar Pokémon
    elif opcion == "3":
        if pokemones:
            for i, p in enumerate(pokemones):
                print(f"{i+1}. {p[0]}")
            num = int(input("¿Cuál deseas eliminar? ")) - 1

            if 0 <= num < len(pokemones):
                pokemones.pop(num)
                print(" Pokémon eliminado del equipo.")
            else:
                print(" Número inválido.")
        else:
            print("No hay pokemones para eliminar.")

    # 4. Actualizar info (tipo o nivel)
    elif opcion == "4":
        if pokemones:
            for i, p in enumerate(pokemones):
                print(f"{i+1}. {p[0]} (Tipo: {p[1]}, Nivel: {p[2]})")

            num = int(input("¿Qué Pokémon deseas actualizar? ")) - 1

            if 0 <= num < len(pokemones):
                print("¿Qué deseas actualizar?")
                print("1. Tipo")
                print("2. Nivel")
                sub = input("Elige opción: ")

                if sub == "1":
                    nuevo_tipo = input("Nuevo tipo: ")
                    pokemones[num][1] = nuevo_tipo
                    print("Tipo actualizado.")

                elif sub == "2":
                    nuevo_nivel = int(input("Nuevo nivel: "))
                    pokemones[num][2] = nuevo_nivel
                    print(" Nivel actualizado.")

                else:
                    print(" Opción inválida.")
            else:
                print("Número inválido.")
        else:
            print("No hay pokemones para actualizar.")

    # 5. Salir
    elif opcion == "5":
        print("¡Gracias por usar tu Pokédex personal! ")
        break

    else:
        print("Opción no válida, intenta nuevamente.")
```
