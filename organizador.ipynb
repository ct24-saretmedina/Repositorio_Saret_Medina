```python
print("===  ORGANIZADOR DE PELÍCULAS FAVORITAS  ===")

peliculas = []  # Cada película será: ["Nombre", puntuación]

while True:
    print("\nMenú de opciones:")
    print("1. Ver lista de películas")
    print("2. Agregar una película")
    print("3. Eliminar una película")
    print("4. Actualizar puntuación de una película")
    print("5. Salir")

    opcion = input("Elige una opción: ")

    # 1. Ver películas
    if opcion == "1":
        if peliculas:
            print("\n Tu lista de películas:")
            for i, p in enumerate(peliculas):
                print(f"{i+1}. {p[0]} -  {p[1]}/10")
        else:
            print("\nNo tienes películas registradas.")

    # 2. Agregar película
    elif opcion == "2":
        nombre = input("Nombre de la película: ")
        puntuacion = float(input("Puntuación (1-10): "))
        peliculas.append([nombre, puntuacion])
        print("Película agregada.")

    # 3. Eliminar película
    elif opcion == "3":
        if peliculas:
            for i, p in enumerate(peliculas):
                print(f"{i+1}. {p[0]}")
            num = int(input("¿Cuál deseas eliminar? ")) - 1
            if 0 <= num < len(peliculas):
                peliculas.pop(num)
                print(" Película eliminada.")
            else:
                print(" Número inválido.")
        else:
            print("No hay películas para eliminar.")

    # 4. Actualizar puntuación
    elif opcion == "4":
        if peliculas:
            for i, p in enumerate(peliculas):
                print(f"{i+1}. {p[0]} -  {p[1]}")
            num = int(input("¿A cuál película cambiarle la puntuación? ")) - 1
            if 0 <= num < len(peliculas):
                nueva = float(input("Nueva puntuación: "))
                peliculas[num][1] = nueva
                print(" Puntuación actualizada.")
            else:
                print(" Número inválido.")
        else:
            print("No hay películas en la lista.")

    # 5. Salir
    elif opcion == "5":
        print("¡Gracias por usar el organizador!")
        break

    else:
        print(" Opción no válida, intenta nuevamente.")
```
