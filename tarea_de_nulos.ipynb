```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

df = pd.read_csv("Data_Science_jobs.csv")

print("\n--- DATASET ORIGINAL ---")
print(df.head())

print("\n--- Columnas con valores nulos ---")
print(df.isnull().any())

print("\n--- Cantidad de nulos por columna ---")
print(df.isnull().sum())

porcentaje_nulos = (df.isnull().mean() * 100).round(2)
print("\n--- Porcentaje de nulos por columna ---")
print(porcentaje_nulos)

print("\n--- Manejo automático de nulos ---")

for col in df.columns:
    if porcentaje_nulos[col] > 40:
        print(f"Eliminando columna '{col}' (más del 40% de nulos)")
        df.drop(columns=[col], inplace=True)
    else:
        if df[col].dtype in ['int64', 'float64']:
            df[col].fillna(df[col].mean(), inplace=True)
            print(f"Rellenando columna numérica '{col}' con la media")
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)
            print(f"Rellenando columna categórica '{col}' con el modo")

print("\n--- DATASET LIMPIO ---")
print(df.head())

df.to_csv("Data_Science_jobs_limpio.csv", index=False)
print("\nArchivo limpio guardado como: Data_Science_jobs_limpio.csv")
```
