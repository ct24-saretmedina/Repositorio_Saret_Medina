```python
print("=== MI TARJETA PERSONAL ===")

# Información personal
nombre = "Saret"
edad = 16
ciudad = "Panamà Oeste,La Chorrera"
mascota ="Chihuahua"

# Cálculo de año de nacimiento
año_actual = 2025
año_nacimiento = año_actual - edad

# Información de mi familia
hermanos = 1
padres = 1

# Operaciones matemáticas
años_para_18 = 18 - edad
años_para_30 = 30 - edad
familia_total = hermanos + padres + 1  # +1 tú

# Mi tarjeta completa
print("Nombre:", nombre)
print("Edad:", edad, "años")
print("Ciudad:", ciudad)
print("Mi mascota es un:",mascota)
print("Año de nacimiento:", año_nacimiento)

print("\n--- Mi Familia ---")
print("Número de hermanos:", hermanos)
print("Padres:", padres)
print("Personas en mi familia incluyendo yo:", familia_total)

print("\n--- Datos Extras ---")
print("Años para cumplir 18:", años_para_18)
print("Años para cumplir 30:", años_para_30)
```
